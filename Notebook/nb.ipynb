```python
from infomax_index import get_index_list, get_index_pdf
lst = get_index_list()
bbig = lst[lst.kr_name.str.contains("KRX BBIG 지수")]
bbig_code = bbig.code.values[0]
bbig_pdf = get_index_pdf(bbig_code)

print(bbig)
```

```python
from infomax_equity import get_equity_daily, get_equity_base
from infomax_index import get_index_daily, get_index_base_info

codes = ",".join(bbig_pdf.code.tolist())
index_code = bbig.code.iloc[0]

dt0 = "20240423"
dt1 = "20240424"
dt2 = "20240425"

index_dt1 = get_index_base_info(index_code, dt1)
index_dt2 = get_index_base_info(index_code, dt2)
index_hist = get_index_daily(index_code, "20240423", dt2)

equity_dt0 = get_equity_base(codes, dt0)
equity_dt1 = get_equity_base(codes, dt1)
equity_dt2 = get_equity_base(codes, dt2)
```

```python
def get_index_change_rate(pdf_df):
    return pdf_df['change_rate'].sum()/len(pdf_df)


print(get_index_change_rate(equity_dt0))
print(get_index_change_rate(equity_dt1))
print(get_index_change_rate(equity_dt2))
```

```python
def get_market_cap(df):
    df['normalized_shares'] = df['listed_shares'] / 100000000.
    df['market_cap'] = df['normalized_shares'] * df['close_price']
    total_cap = df['market_cap'].sum()
    df['market_cap_ratio'] = df['market_cap'] / total_cap

    return df

res0 = get_market_cap(equity_dt0)
res1 = get_market_cap(equity_dt1)
res2 = get_market_cap(equity_dt2)
```
